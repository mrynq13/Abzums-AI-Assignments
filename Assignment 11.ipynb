```python
import requests

BASE_URL = "https://mastermind.darkube.app"

# Step 2: A Function to Create a Game
def start_game():
    response = requests.post(f"{BASE_URL}/game")
    data = response.json()
    return data["game_id"]
```

```python

#  Step 3: A Function to Submit Guesses to the API
def send_guess(game_id, guess):
    response = requests.post(
        f"{BASE_URL}/guess",
        json={"game_id": game_id, "guess": guess}
    )
    return response.json()
```

```python

#  Step 4: A Function to Prompt Guess
def prompt_guess(game_id):
    while True:
        guess = input("Enter your guess: ")

        # Check: length must be 4
        if len(guess) != 4:
            print("Guess must be 4 characters long.")
            continue

        # Check: only digits 1–6
        if not all(d in "123456" for d in guess):
            print("Guess must only contain digits 1–6.")
            continue

        # Check: all digits unique
        if len(set(guess)) != 4:
            print("Digits must be unique.")
            continue

        result = send_guess(game_id, guess)
        print(f"Bulls: {result['bulls']}, Cows: {result['cows']}")

        # Check for win
        if result["bulls"] == 4:
            print("You win!")
            break
```

```python

#  Step 5: A Function to Run the Game
def run_game():
    print("Welcome to Mastermind!")
    game_id = start_game()
    prompt_guess(game_id)

# Run the game
if __name__ == "__main__":
    run_game()
```
